```python
import pandas as pd #importar la libreria pandas

df = pd.read_csv("figuras_geometricas.csv") #df es la variable donde guardo el dataframe
df.head() #para ver las primeras filas y chequear que se cargo bien
```

```python
df.info()

df.isnull().sum() #aca chequeo que hay 3 figuras nulas y 6 colores nulos
```

```python
df['figura'].fillna("Sin figura") #fillna es para rellenar
```

```python
df.isnull().sum() #aca sigue saliendo 3 figuras nulas, tengo que hacer el .fillna en un nuevo df
```

```python
df['figura'] = df['figura'].fillna("Sin figura") #haciendo asi, llena los valores en el dataframe original
df.isnull().sum() #se ve que ya no hay mas nulos
```

```python
df['color'] = df['color'].fillna("Sin color")
df.isnull().sum()
```
